```python
import schemdraw as sd
import schemdraw.elements as e
```

```python
d = sd.Drawing(unit=5)
V1 = d.add(e.SourceV(label='$20V$'))
R1 = d.add(e.Resistor(d='right', label='400$\Omega$'))
d.add(e.Dot())
d.push()
R2 = d.add(e.Resistor(d='down', botlabel='100$\Omega$', lblrotate=True))
d.add(e.Dot())
d.pop()
L1 = d.add(e.Line())
I1 = d.add(e.SourceI(d='down', botlabel='1A'))
L2 = d.add(e.Line(d='left', tox=V1.start))
d.loopI([R1,R2,L2,V1], '$I_1$', pad=1.25)
d.loopI([R1,I1,L2,R2], '$I_2$', pad=1.25)  # Use R1 as top element for both so they get the same height
d.draw()
```

```python
d = sd.Drawing(unit=5,lw=1)
V1 = d.add(e.SourceV(label=r'$20\,\mathrm{V}$',lblofst=0.2))
L1 = d.add(e.Inductor2(loops=2,d='right', label=r'$400\,\mathrm{mH}$'))
d.add(e.Dot())
d.push()
R2 = d.add(e.Resistor(d='down', botlabel=r'$100\,\Omega$', lblrotate=True))
d.add(e.Dot())
d.pop()
L1 = d.add(e.Line())
I1 = d.add(e.SourceI(d='down', botlabel=r'$1\,\mathrm{A}\angle 22^{\circ}$',lblofst=0.2))
L2 = d.add(e.Line(d='left', tox=V1.start))
d.loopI([R1,R2,L2,V1], '$I_1$', pad=1.25)
d.loopI([R1,I1,L2,R2], '$I_2$', pad=1.25)  # Use R1 as top element for both so they get the same height
d.draw()
```
